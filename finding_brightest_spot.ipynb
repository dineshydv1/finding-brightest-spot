## finding brightest spot in image



```python
# import the necessary packages
import numpy as np
import argparse
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# define input arguments
# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--image", help="Path to the image file")
# ap.add_argument("-r", "--radius", type=int, help="radius of Gaussian blur; \
#                 must be odd")
# args = vars(ap.parse_args())

# arguments above are commented because we are running from jyputer notebook
# need to enable them if running as a script and disable below assignment
# Defined value to arguments below

image = 'retina.png' # Path to the image file
radius = 3  # radius of Gaussian blur- must be odd
```

```python
# read image and convert to gray
image = cv2.imread(image)
orig = image.copy()
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# show image
#cv2.imshow('moon',image)
#cv2.waitKey(0)

# plt code
print(image.shape)
plt.imshow(image)
```

```python
# show gray image
# cv2.imshow('moon',gray)
# cv2.waitKey(0)

# plt code
print(gray.shape)
plt.imshow(gray)
```

```python
# minMaxLoc does not work with multi-channel arrays
# thats why we converted image to gray scale
# it gives minimum and maximum values and their positions

(minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(gray)
print("(minVal, maxVal, minLoc, maxLoc):",(minVal, maxVal, minLoc, maxLoc))

image2 = image.copy()
# draw circle on image
cv2.circle(image2,   # image on which to draw the circle
           maxLoc,   # center coordinate 
           10,       # radius in pixels
           (255, 0, 0), # radius color # red
           2)        # radius thickness
#cv2.imshow("Naive", image)
#cv2.waitKey(0)

plt.imshow(image2)
```

```python

#gray = cv2.GaussianBlur(gray, (args["radius"], args["radius"]), 0)
gray = cv2.GaussianBlur(gray, (radius, radius), 0)
plt.imshow(gray)

(minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(gray)
print("(minVal, maxVal, minLoc, maxLoc):",(minVal, maxVal, minLoc, maxLoc))
```

```python
image = orig.copy()
cv2.circle(image, maxLoc, radius, (255, 0, 0), 2)

# cv2.imshow("Robust", image)
# cv2.waitKey(0)

plt.imshow(image)
```
